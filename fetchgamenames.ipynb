```python
import pandas as pd
import steamspypi
pd.options.display.max_rows = 986
```

```python
df = pd.read_pickle('testingdf2.pkl')
```

```python
df
```

```python
spy_request=dict()
spy_request['request']='appdetails'
```

```python
gamenames=[]

i=1
for gameid in df.index:
    print(i)
    spy_request['appid']=gameid
    spy_return=steamspypi.download(spy_request)
    gamenames.append(spy_return['name'])
    i+=1
```

```python
price=[]

i=1
for gameid in df.index:
    print(i)
    spy_request['appid']=gameid
    spy_return=steamspypi.download(spy_request)
    price.append(spy_return['price'])
    i+=1
```
